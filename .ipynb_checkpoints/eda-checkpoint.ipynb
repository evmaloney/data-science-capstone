# Data Science Capstone, Part 2: Brief

## Dataset Overview

### Vehicle Fuel Economy Estimates, 1984-2017
#### Main columns of use:
- **Vehicle ID** - Primary key for specific vehicle.
- **Year** - Year vehicle was made.
- **Make** - Make of vehicle.
- **Model** - Model of vehicle.
- **Class** - Size/Body Type of vehicle.
- **Drive** - 2-Wheel vs 4-Wheel Drive of vehicle.
- **Transmission** - Automatic vs Manual Transmission of vehicle.
- **Engine Cylinders** - No. of Cylinders in vehicle.
- **Fuel Type** - Type of Fuel vehicle takes.
- **City MPG (FT1)** - City Miles per Gallon for vehicle.
- **Highway MPG (FT1)** - Highway Miles per Gallon for vehicle.
- **Tailpipe CO2 in Grams/Mile (FT1)** - Carbon emissions in Grams per Mile for vehicle.

```python
import pandas as pd
%matplotlib inline
```

```python
db = pd.read_csv("database.csv")
```

```python
db.head()
```

```python
db.shape
```

```python
db.dtypes
```

### Shrink data to columns of interest

```python
cars = db[['Vehicle ID', 'Year', 'Make', 'Model', 'Class', 'Drive', 'Transmission', 'Engine Cylinders', 'Fuel Type', 'City MPG (FT1)', 'Highway MPG (FT1)', 'Tailpipe CO2 in Grams/Mile (FT1)']]
cars
# Vehicle ID - Primary key for specific vehicle.
# Year - Year vehicle was made.
# Make - Make of vehicle.
# Model - Model of vehicle.
# Class - Size/Body Type of vehicle.
# Drive - 2-Wheel vs 4-Wheel Drive of vehicle.
# Transmission - Automatic vs Manual Transmission of vehicle.
# Engine Cylinders - No. of Cylinders in vehicle.
# Fuel Type - Type of Fuel vehicle takes.
# City MPG (FT1) - City Miles per Gallon for vehicle.
# Highway MPG (FT1) - Highway Miles per Gallon for vehicle.
# Tailpipe CO2 in Grams/Mile (FT1) - Carbon emissions in Grams per Mile for vehicle.
```

### Rename a few columns

```python
cars = cars.rename(columns={'City MPG (FT1)': 'City MPG'}).rename(columns={'Highway MPG (FT1)': 'Highway MPG'}).rename(columns={'Tailpipe CO2 in Grams/Mile (FT1)': 'CO2 g/mi'})
cars
```

### View just the Toyota data

```python
cars_toyota = cars[cars['Make'] == 'Toyota']
cars_toyota
```

### View just the Honda data

```python
cars_honda = cars[cars['Make'] == 'Honda']
cars_honda
```

### View just the Toyota Camry data

```python
cars_toyota_camry = cars_toyota[cars_toyota['Model'] == 'Camry']
cars_toyota_camry
```

### View just the Honda Accord data

```python
cars_honda_accord = cars_honda[cars_honda['Model'] == 'Accord']
cars_honda_accord
```

### Edit workable data to just Accords and Camrys

```python
cars = cars[(cars['Make'] == 'Honda') & (cars['Model'] == 'Accord') | (cars['Make'] == 'Toyota') & (cars['Model'] == 'Camry')]
cars
```

### Get a rough look at Camry and Accord CO2 output by mean

```python
cars_toyota_camry['CO2 g/mi'].mean()
```

```python
cars_honda_accord['CO2 g/mi'].mean()
```

### Examine any NaN values and replace them

```python
missing_values = pd.DataFrame(cars.isnull().sum(), columns=['Missing Values'])
missing_values
```

```python
nan_values = cars[cars.isnull().any(axis=1)]
nan_values
```

```python
cars_honda_accord['Drive'].value_counts()
```

```python
cars_toyota_camry['Drive'].value_counts()
```

In research, all NaN values for these cars should be 'Front-Wheel Drive'

```python
cars['Drive'] = cars['Drive'].fillna(value='Front-Wheel Drive')
cars['Drive'].value_counts()
```

### Plot each car separately and together to see how they relate

```python
accord_plot = cars_honda_accord.plot(kind="scatter", x="Year", y="CO2 g/mi", color="Red", label="Honda Accord CO2 g/mi Output by Year");
```

```python
camry_plot = cars_toyota_camry.plot(kind="scatter", x="Year", y="CO2 g/mi", color="Green", label="Toyota Camry CO2 g/mi Output by Year");
```

```python
accord_plot = cars_honda_accord.plot(kind="scatter", x="Year", y="CO2 g/mi", color="Red", label="Honda Accord CO2 g/mi Output by Year");
cars_toyota_camry.plot(kind="scatter", x="Year", y="CO2 g/mi", color="Green", label="Toyota Camry CO2 g/mi Output by Year", ax=accord_plot);
```

#### After going over the data to this EDA, overall, it looks like both cars emissions have an uptick in the 1990s and 2000s and then come down in the 2010s. In general, they both look like they ended better than they started. The Camry had a better CO2 output in the 80s. Going into the 90s, the Accord stayed pretty steady while the Camry's emmissions rose and the Accord generally stayed lower than the Camry into the 2000s and 2010s. Using the eyeball test, the Accord has done a better job overall and curbing its CO2 emissions. While the Camry did improve post 90s, the Accord generally stayed below the Camry and is the better choice going forward.
